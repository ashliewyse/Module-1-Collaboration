### 3.2
Assign the result to a variable called seconds_per_hour.


```python
seconds_per_hour = 60 * 60
seconds_per_hour
```

```python
seconds_per_hour * 24
```

```python
seconds_per_day = seconds_per_hour * 24
seconds_per_day
```

```python
seconds_per_day / seconds_per_hour
```

```python
seconds_per_day // seconds_per_hour
```

Yes, the integer division result (24) agrees with the floating-point result (24.0) except for the .0 at the end.
