File: deans_list_honor_roll.ipynb  
Description: This program accepts student names and GPAs and determines
whether each student qualifies for the Dean's List or the Honor Roll.


```python

while True:
    last_name = input("Enter student's last name (or ZZZ to quit): ").strip()

    if last_name.upper() == "ZZZ":
        print("Exiting program.")
        break

    first_name = input("Enter student's first name: ").strip()
    gpa = float(input("Enter student's GPA: "))

    if gpa >= 3.5:
        print(first_name, last_name, "has made the Dean's List.")
    elif gpa >= 3.25:
        print(first_name, last_name, "has made the Honor Roll.")
    else:
        print(first_name, last_name, "did not qualify for honors.")
```
