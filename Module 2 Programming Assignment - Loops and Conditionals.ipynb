## 4.1
Choose a number for secret and guess and compare them using if, elif, and else.


```python
secret = 7
guess = 4

if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

## 4.2
Assign True/False to small and green. Use if statements to print which item matches:
cherry, pea, watermelon, pumpkin.


```python
small = True
green = False

# cherry: small True, green False
if small and not green:
    print("cherry")

# pea: small True, green True
if small and green:
    print("pea")

# watermelon: small False, green True
if not small and green:
    print("watermelon")

# pumpkin: small False, green False
if not small and not green:
    print("pumpkin")
```

## 6.2
Use a while loop to compare number to guess_me and print:
"too low", "found it!", or "oops".


```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break

    number += 1
```

## 6.3
Use a for loop over range(10). Print "too low" if number < guess_me,
"found it!" if equal (then break), and "oops" if greater (then exit loop).


```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
```
